# O método de Monte Carlo
 

O método de monte carlo consiste em criar uma quantidade finita de pontos aleatóriamente dentro um intervalo dado e definir o que, por exemplo, para calcular a área de um círculo inscreveriamos um circulo (PONTOS VERMELHOS) em um quadrado (INTERVALO/PONTOS AZUIS) e criariamos uma quantidade N de pontos em posições aleatoriamente definidas, alguns desses pontos estariam no quadrado e outros no círculo, os pontos dentro do quadrado seriam descartados e os que estão no círculo seriam somados para assim acharmos a área do círculo.

# Entendendo o algoritmo

A intuito de exemplificar o funcionamento do método de monte carlo em python, utilizei a seguinte função:
$y =  0.04*20*1.2^{(x-1)}$
Iremos calcular a área abaixo dessa curva. A priori, é necessário definirmos nossa função IMPLICITAMENTE, isto é, quando as variáveis utilizadas estão em apenas um lado da equação, ou melhor, formalmente, temos que $R(x,f(x)) = 0$.
Portanto, temos: $ y - 0.04*20*1.2^{(x-1)}= 0$


imagem2.png

Como podemos ver acima o plot da função. Sendo assim, comecemos a codar.

```python
import random

TOTAL_PONTOS = 1000
#a = int(input("Intervalo a: \t")) #Utilizaremos o intervalo [0,10]
#b = int(input("Intervalo b: \t"))
a = 10
b = 50
integral = 0

f = lambda x: 0.04*20*1.2**(x-1) #Definimos a função
for i in range(TOTAL_PONTOS): #
    x = random.uniform(a,b) #Criamos um par de pontos a cada iteração de distância uniforme aleatoriamente
    y = random.uniform(a,b)
    implicit_func = y - f(x) #Aqui é a definição de função implicita
    if(implicit_func <= 0): #Se o valor da função implicita for menor que zero, então esse ponto está ABAIXO DA CURVA
        integral += f(x) #Armazenamos cada ponto abaixo da curva e somamos.
elem_infinitesimal = (b-a)/TOTAL_PONTOS
area = elem_infinitesimal*integral # b-a é a variação de x, ou seja, o delta x para cada ponto. Isto serve para compensar os valores dado pela integral
#Ou melhor dizendo, a definição de integral leva-se em conta um dx, o elemento infinitesimal.
area
```

# Observações

Por mais que esse método dê um valor razoavelmente bom de estimativa, vale lembrar que o algoritmo utilizado não levou em conta os ERROS DE ARRENDODAMENTO e ERROS DE TRUNCAMENTO, isto fará com que em valores de intervalo que são muito próximos o ERRO seja grande, em comparação com intervalos bem mais distantes. Nesse, caso pode-se observar que a quantidade de iterações no método de riemann é menor que a de Monte Carlo. A propósito, Monte Carlo só é utilizado em soluções númericas de integrais muito complexas que nenhum outro método resolve.
