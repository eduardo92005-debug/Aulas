# O mtodo de Monte Carlo
 

O mtodo de monte carlo consiste em criar uma quantidade finita de pontos aleatriamente dentro um intervalo dado e definir o que, por exemplo, para calcular a rea de um crculo inscreveriamos um circulo (PONTOS VERMELHOS) em um quadrado (INTERVALO/PONTOS AZUIS) e criariamos uma quantidade N de pontos em posies aleatoriamente definidas, alguns desses pontos estariam no quadrado e outros no crculo, os pontos dentro do quadrado seriam descartados e os que esto no crculo seriam somados para assim acharmos a rea do crculo.

# Entendendo o algoritmo

A intuito de exemplificar o funcionamento do mtodo de monte carlo em python, utilizei a seguinte funo:
$y =  0.04*20*1.2^{(x-1)}$
Iremos calcular a rea abaixo dessa curva. A priori,  necessrio definirmos nossa funo IMPLICITAMENTE, isto , quando as variveis utilizadas esto em apenas um lado da equao, ou melhor, formalmente, temos que $R(x,f(x)) = 0$.
Portanto, temos: $ y - 0.04*20*1.2^{(x-1)}= 0$


imagem2.png

Como podemos ver acima o plot da funo. Sendo assim, comecemos a codar.

```python
import random

TOTAL_PONTOS = 1000
#a = int(input("Intervalo a: \t")) #Utilizaremos o intervalo [0,10]
#b = int(input("Intervalo b: \t"))
a = 10
b = 50
integral = 0

f = lambda x: 0.04*20*1.2**(x-1) #Definimos a funo
for i in range(TOTAL_PONTOS): #
    x = random.uniform(a,b) #Criamos um par de pontos a cada iterao de distncia uniforme aleatoriamente
    y = random.uniform(a,b)
    implicit_func = y - f(x) #Aqui  a definio de funo implicita
    if(implicit_func <= 0): #Se o valor da funo implicita for menor que zero, ento esse ponto est ABAIXO DA CURVA
        integral += f(x) #Armazenamos cada ponto abaixo da curva e somamos.
elem_infinitesimal = (b-a)/TOTAL_PONTOS
area = elem_infinitesimal*integral # b-a  a variao de x, ou seja, o delta x para cada ponto. Isto serve para compensar os valores dado pela integral
#Ou melhor dizendo, a definio de integral leva-se em conta um dx, o elemento infinitesimal.
area
```
