# **Entendendo algoritmos**

> Imagine a seguinte situação: Um alien da galáxia de Sombrero chamado Jackson criou um código para analisar uma lista imensa de números, você curioso como é analisou o algoritmo e concluiu que é um algoritmo de busca, porém você percebeu que esse algoritmo é extremamente ineficiente e ele demora demais para achar a entrada dada. Como você mostraria para o alien que o algoritmo dele é ruim? Toma-se como hipotése que a matemática é uma linguagem universal e qualquer pessoa em qualquer lugar do universo entenderia.

Para esse tipo de situação que utilizamos algo que vamos chamar de análise de complexidade de algoritmo, ela basicamente é dividida em duas a complexidade temporal e a de espaço, isto é, o quanto seu algoritmo demora para resolver um dado problema e o quanto de espaço ele tem que alocar na memória para resolver esse problema. A priori, vamos adentrar na complexidade temporal e definir ela de uma forma bem amigável.
Basicamente, pense da seguinte forma, todo algoritmo pode ser analisado através de um tempo de processamento, isto é, o tempo em relação a quantidade de processamento ou operações que seu computador precisa fazer, e isso podemos definir como uma função matemática. Assim, analisando e comparando a eficiência do seu algoritmo com o de outros, você consegue concluir se seu algoritmo é bom ou não.
